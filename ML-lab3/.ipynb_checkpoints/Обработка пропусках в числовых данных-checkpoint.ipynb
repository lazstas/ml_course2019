```python
# Выберем числовые колонки с пропущенными значениями
# Цикл по колонкам датасета
num_cols = []
for col in data.columns:
    # Количество пустых значений 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0 and (dt=='float64' or dt=='int64'):
        num_cols.append(col)
        temp_perc = round((temp_null_count / total_count) * 100.0, 2)
        print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

```python
# Фильтр по колонкам с пропущенными значениями
data_num = data[num_cols]
data_num
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
```

```python
data_num_HP = data_num[['Engine HP']]
data_num_HP.head()
```

```python
# Фильтр для проверки заполнения пустых значений
indicator = MissingIndicator()
mask_missing_values_only = indicator.fit_transform(data_num_HP)
mask_missing_values_only
```

```python
strategies=['mean', 'median','most_frequent']
```

```python
def test_num_impute(strategy_param):
    imp_num = SimpleImputer(strategy=strategy_param)
    data_num_imp = imp_num.fit_transform(data_num_HP)
    return data_num_imp[mask_missing_values_only]
```

```python
strategies[0], test_num_impute(strategies[0])
```

```python
strategies[1], test_num_impute(strategies[1])
```

```python
strategies[2], test_num_impute(strategies[2])
```

```python
# Более сложная функция, которая позволяет задавать колонку и вид импьютации
def test_num_impute_col(dataset, column, strategy_param):
    temp_data = dataset[[column]]
    
    indicator = MissingIndicator()
    mask_missing_values_only = indicator.fit_transform(temp_data)
    
    imp_num = SimpleImputer(strategy=strategy_param)
    data_num_imp = imp_num.fit_transform(temp_data)
    
    filled_data = data_num_imp[mask_missing_values_only]
    
    return column, strategy_param, filled_data.size, filled_data[0], filled_data[filled_data.size-1]
```

```python
data[['Engine HP']].describe()
```

```python
test_num_impute_col(data, 'Engine HP', strategies[0])
```

```python
test_num_impute_col(data, 'Engine HP', strategies[1])
```

```python
test_num_impute_col(data, 'Engine HP', strategies[2])
```
